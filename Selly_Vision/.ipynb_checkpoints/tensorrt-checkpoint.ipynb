```python
import tensorflow as tf
import numpy as np
import cv2
import time
from model.pspunet import pspunet
import matplotlib.pyplot as plt
from util.path import create_mask

gpus = tf.config.experimental.list_physical_devices('GPU')

if gpus:
    try:
        tf.config.experimental.set_virtual_device_configuration(
       gpus[0],
        [tf.config.experimental.VirtualDeviceConfiguration(memory_limit=3000)])
    except RuntimeError as e:
        print(e)


cap = cv2.VideoCapture("test2.mp4")

i=0
while i<130:
    _, frame = cap.read()
    frame = cv2.resize(frame, (480, 272))
    frame = frame[tf.newaxis , ...]
    frame = np.array(frame, dtype=np.float32)
    i+=1

            
IMG_WIDTH = 480
IMG_HEIGHT = 272
n_classes = 7

model = pspunet((IMG_HEIGHT, IMG_WIDTH ,3), n_classes)
model.load_weights("pspunet_weight.h5")



loaded = tf.saved_model.load("./trt_fp16")
print(list(loaded.signatures.keys()))  # ["serving_default"]



infer = loaded.signatures["serving_default"]
print(infer.structured_outputs)

#tensorrt model
trt = infer(tf.convert_to_tensor(frame))
trt = trt['conv2d_37'].numpy()

#tensorflow model
tensor = model.predict(tf.convert_to_tensor(frame))

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.axis("off")
plt.title("tensorrt")
plt.imshow(tf.keras.preprocessing.image.array_to_img(create_mask(trt)))
plt.subplot(1,2,2)
plt.axis("off")
plt.title("tensorflow")
plt.imshow(tf.keras.preprocessing.image.array_to_img(create_mask(tensor)))

```

```python
start_time = time.time()
for i in range(100):
    trt = infer(tf.convert_to_tensor(frame))
    trt = trt['conv2d_37'].numpy()
    
trt_time = (time.time()-start_time)/100

start_time = time.time()

for i in range(100):
    tensor = model.predict(tf.convert_to_tensor(frame))

tensor_time = (time.time()-start_time)/100

print("tensorrt : ", round(trt_time, 3)) 
print("tensorflow : ", round(tensor_time, 3))
      
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.axis("off")
plt.title("tensorrt FP16, time : " + str(round(trt_time, 4)))
plt.imshow(tf.keras.preprocessing.image.array_to_img(create_mask(trt)))

loaded = tf.saved_model.load("./trt_fp32")
print(list(loaded.signatures.keys()))  # ["serving_default"]


infer = loaded.signatures["serving_default"]
print(infer.structured_outputs)

for i in range(100):
    trt = infer(tf.convert_to_tensor(frame))
    trt = trt['conv2d_37'].numpy()
    
trt_time = (time.time()-start_time)/100

start_time = time.time()

plt.subplot(1,3,2)
plt.axis("off")
plt.title("tensorrt FP32, time : " + str(round(trt_time, 4)))
plt.imshow(tf.keras.preprocessing.image.array_to_img(create_mask(trt)))
plt.subplot(1,3,3)
plt.axis("off")
plt.title("tensorflow, time : " + str(round(tensor_time, 4)))
plt.imshow(tf.keras.preprocessing.image.array_to_img(create_mask(tensor)))

```
